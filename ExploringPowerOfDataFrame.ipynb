<h5>String handling(): </h5>

```python
# Select all rows where the column starts with the string.

df[df['Name'].str.match('Jau')]

# Select all rows where the column contain a string.

df[df['Name'].str.contains('singh')]
```

<h5>Applying a function on DataFrame: </h5>

We can apply a function on the column of the DataFrame.

```python
df[df["Weight"] > 0].head().Weight
```

```python
df[df["Weight"] > 0].head().Weight * 3
```

```python
# Create a function
def multiply2(x): 
    num = x
    num *= 2
    return num

# Apply a function
df[df["Weight"] > 0].head().Weight.apply(multiply2)
```

```python
# Apply anonymous function
df[df["Weight"] > 0].head().Weight.apply(lambda x: x * 4)
```
