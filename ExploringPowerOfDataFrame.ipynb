### Manage Columns, Merging & Joining:


<i>Create two DataFrame from sample dataset. We will use these throughout this section.</i>

```python
# Create a sample DataFrame for this section:
# Employee dataset one.
dataSet1 = {
    'EID': [1001, 1002, 1003, 1004, 1005], 
    'Name': ['Dhoni', 'Virat', 'Sachin', 'Laxman', 'Saurav'],
    'Salary': [45000, 55000, 60000, 35000, 40000]
}

emp_df1 = pd.DataFrame(dataSet1)
emp_df1
```

```python
# Employee dataset two.
dataSet2 = {
    'EID': [1001, 1002, 1003, 1004, 1005, 1008], 
    'City': ['Ranchi', 'Delhi', 'Mumbai', 'Hyderabad', 'Kolkata', 'BBSR']
}

emp_df2 = pd.DataFrame(dataSet2)
emp_df2
```

<h5>Add Column in DataFrame:</h5>

```python
# Add a new column in DataFrame from a fresh list.
emp_df1["Age"] = [39, 31, 45, 41, 46]

# Add a new column in DataFrame from existing column.
emp_df1["AgeTwice"] = emp_df1["Age"] * 2

# Add a new column in DataFrame from existing two columns.
emp_df1["Sum"] = emp_df1["Age"] + emp_df1["AgeTwice"] 

# Add a new column in DataFrame from existing column with applying a function.
emp_df1["SumSq"] = emp_df1["Sum"].apply(lambda x: x*x)

emp_df1
```

<h5>Delete Column in DataFrame:</h5>

Column deletion in DataFrame using drop function. It has 2 parameters axis and inplace.<br />
<ul>
    <li><b>axis=1</b> implies delete column wise,</li>
    <li><b>axis=0</b> implies delete rowwise</li>
    <li><b>inplace=True</b> implies Modify the DataFrame</li>
</ul>

```python
# Drop a column.
emp_df1.drop('AgeTwice', axis=1, inplace=True) 

# You can drop a row as well, using below statement. 
#new_people_df.drop('Summer', axis=0, inplace=True) 

# Drop multiple columns
emp_df1.drop(['Sum', 'SumSq'], axis=1, inplace=True) 

emp_df1
```

<h5>Concatenating two Dataframes:</h5>


```python
# Employee dataset three.
dataSet3 = {
    'EID': [1006, 1007], 
    'Name': ['Sunil', 'kapil'],
    'Salary': [45000, 46000]
}

emp_df3 = pd.DataFrame(dataSet3)

print(emp_df1)
print('\n\n')
print(emp_df3)
```

```python
# Concate dataframe along the rows
emp_df_new = pd.concat([emp_df1, emp_df3], axis=0)
emp_df_new
```

```python
# Concate dataframe along the columns
emp_df_new = pd.concat([emp_df1, emp_df3], axis=1)
emp_df_new
```

<h5>Merging dataFrames</h5>

Merge two dataframes along the EID value

```python
print(emp_df1)
print('\n\n')
print(emp_df2)
```

```python
emp_df_new = pd.merge(emp_df1, emp_df2, on='EID')
emp_df_new
```

```python
emp_df_new = pd.merge(emp_df1, emp_df2, on='EID', how='outer')

emp_df_new
```

<h5>Joining DataFrames:</h5>


Index must needed to perform Join in pandas.

```python
left = emp_df1.set_index("EID")
right = emp_df2.set_index("EID")

# Index must needed to perform Join in pandas.
# Join operates on index, where Merge can be applied over multiple columns.
emp_df_new = left.join(right, how="inner", lsuffix='_l', rsuffix='_r')
emp_df_new
```

### Grouping:

Pandas groupby() function is used to split the data into groups based on some criteria. <br />

By “group by” we are referring to a process involving one or more of the following steps: <br />

<ul>
    <li>Splitting the data into groups based on some criteria.</li>
    <li>Applying a function to each group independently.</li>
    <li>Combining the results into a data structure.</li>
</ul>


```python
# Finding data groupby sport.
dfSport = df.sample(2000).groupby("Sport")

# Also can groupby multiple columns.
# df.groupby(['Sport', 'Event']) 

# Print the type of groupby return data.
type(dfSport)
```

```python
# Let's print the first entries in all the groups formed. 
dfSport.first().head()
```

```python
# Finding the values contained in the "Athletics" group in Sport
dfSportAth = dfSport.get_group('Athletics')

# Find the first five gold medalist.
dfSportAth[dfSportAth.Medal == "Gold"].head()
```

```python
# dfSportAth.count()
dfSportAth.mean()
```

```python
dfSportAth['Weight'].sum()
```

Applying multiple functions to columns in groups

```python
dfSport.agg(
    {
        # find the min, max, and sum of the duration column
        'Age': [min, max, sum],
         # find the number of network type entries
        'ID': "count",
        # min, first, and number of unique dates per group
        'Year': [min, 'first', 'nunique']
    }
).head(10)
```
